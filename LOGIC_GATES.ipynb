# BUILD THE LOGIC GATES WITH NEURAL NETWORK


In this Notebook, we are trying to build the Logic Gates Known as OR GATE, XOR GATE and AND GATE. To build theses gates, we need two entries and one output. The entries are the combination of 0 and 1 and the Output can only be 0 or 1. We will use Numpy and sklearn librairies.

```python
import numpy as np
from sklearn.neural_network import MLPClassifier

import warnings
warnings.filterwarnings('ignore')
```

```python
#create the entries and the output
# the entries can be represented by an array of dimension 2 (4,2)
#the output can be represented by a vector (4,1)

X_OR= np.array([[0,0],[0,1],[1,0],[1,1]])
Y_OR= np.array([0,1,1,1])

#visualisation
print("X is", X_OR)
print("Y is", Y_OR)
```

```python
#create the model
modelOR= MLPClassifier(hidden_layer_sizes=(50,), activation='logistic', solver='adam', max_iter=1000, verbose=10)

#train the model
modelOR.fit(X_OR,Y_OR)

#make a prediction
Y_OR_hate= modelOR.predict(X_OR)
print("The Prediction is given by",Y_OR_hate)

#compute the score of our model
score = modelOR.score(X_OR,Y_OR)
print("The score is",score)
```

#### As we can noticed, Y_OR_hate is the same as Y_OR. It means that the model suceed to predict the OR GATE
